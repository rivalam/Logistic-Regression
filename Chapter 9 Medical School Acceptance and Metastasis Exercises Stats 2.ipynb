## Medical School Acceptance.
The datafile MedGPA used in Example 4 also contains information on the medical school admission test (MCAT)scores
for the same sample of 55 students. Fit a logistic model to predict the Acceptace status using the MCAT scores.

### Write down the estimated versions of both the logit and probability forms of this model.

```r
load("MedGPA.RDA")

model<-glm(Acceptance~MCAT,family=binomial(link='logit'),data=MedGPA)
summary(model)
```

```r
model<-glm(Acceptance~MCAT,family=binomial(link='logit'),data=MedGPA)
summary(model)
```

```r
modell<-glm(Acceptance~GPA,family=binomial(link='logit'),data=MedGPA)
summary(modell)
```

*Fitted model is linear in the log(odds) scale: logit(P(Acceptance))*=

-8.71+0.25MCAT

*Probability form: P(Accept)*= 
$\frac{e^(-8.71+0.25MCAT)}{1+e^(-8.71+0.25MCAT)}$

### What would the estimated model say about the chance that a student with MCAT=40 is accepted to medical school?

```r
log_odds<-(-8.71)+(0.25*40)
exp(log_odds)/(1+exp(log_odds))
```

The chance that a student with MCAT=40 is accepted to medical school is approximately 78.41%

### For approximately what MCAT score would a student have roughly a 50-50 chance of being accepted to medical school? 
(Hint: You might look at a graph or solve one of the equations algebraically)

```r
library(ggplot2)
ggplot(MedGPA, aes(x=MCAT, y=Acceptance)) + geom_point() + 
stat_smooth(method="glm", method.args=list(family="binomial"), se=FALSE)+ggtitle("Logistic Model for MCAT Score vs. Acceptance")
```

We can see from the graph that the acceptance rate is 50-50 when the MCAT score is approximately 36-37.
Algebraically, the MCAT score is approx. 34.84

## Metasizing cancer.
In a study of 31 patients with esophageal cancer, it was found that in 18 of the patients the cancer had metasized to the lymph nodes. Thus, an overall estimate of the probablity of the metastasis is 18/31 = 0.58
A predictor variable measured on each patient is Size of the tumor (in cm). A fitted logistic regression model is 
       log($\frac{pi}{1-pi}$) = -2.086+0.5117(Size)

### A. Use this model to estimate the odds of metastisis, pi/(1-pi), if a patient's tumor size is 6 cm.

```r
Odds_6cm<-exp(-2.086+0.5117*(6))
Odds_6cm
```

The odds of metastisis is 2.68 if a patient's tumor size is 6 cm.

### B. Use the model to predict the probability of metastasis if a patient's tumor is 6 cm.

```r
log_odds_six<-(-2.086)+(0.5117*6)
exp(log_odds_six)/(1+exp(log_odds_six))
```

The probability of metastasis if a patient's tumor is 6cm is 72.79%.

## C. How much do the estimated odds change if the tumor size changes from 6cm to 7 cm? Provide and interpret an odds ratio.

```r
Odds_7cm<-exp(-2.086+0.5117*(7))
Odds_7cm
Odds_7cm/Odds_6cm
```

The odds of getting metasis is 1.67 times higher with the tumor size being 7 cm than being 6 cm.
